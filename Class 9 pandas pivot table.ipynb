```python
import pandas as pd
import numpy as np
```

```python
# Pivot_table
```

```python
df = pd.read_excel(r"C:\Users\itspark\Documents\Analytics\dataset/sales-funnel.xlsx")
```

```python
df
```

```python
pd.pivot_table(df,index=['Name'])
```

```python
#pass multiple indexises
pd.pivot_table(df,index=['Name','Rep','Manager'])
```

```python
pd.pivot_table(df,index=['Manager','Rep'])
```

```python
pd.pivot_table(df,index=['Manager','Rep'],values=['Price'])
```

```python
#aggfunc
pd.pivot_table(df,index=['Manager','Rep'],values=['Price'],aggfunc=[len,np.mean])
```

```python
# sum of price
pd.pivot_table(df,index=['Manager','Rep'],values=['Price'],aggfunc=np.sum)
```

```python
pd.pivot_table(df,index=['Manager','Rep'],values=['Price'],columns=['Product'],aggfunc=[np.sum])
```

```python
df['Product'].unique()
```

```python
pd.pivot_table(df,index=['Manager','Rep'],values=['Price'],columns=['Product'],aggfunc=[np.sum],fill_value = 0)
```

```python
pd.pivot_table(df,index=['Manager','Rep'],values=['Price','Quantity'],columns=['Product'],aggfunc=[np.sum],fill_value = 0)
```

```python
pd.pivot_table(df,index=['Manager','Rep','Product'],values=['Price','Quantity'],aggfunc=[np.sum],fill_value = 0,margins=True)
```

```python
final_table = pd.pivot_table(df,index=['Manager','Status'],columns=['Product'],values=['Quantity','Price'],
               aggfunc={'Quantity':len,'Price':[np.sum,np.mean]},fill_value=0)
```
